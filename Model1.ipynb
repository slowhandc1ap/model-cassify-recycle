```python
from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
```

```python
X =  np.empty([0])
y = np.empty([0])
filename_R = os.listdir(path+"/R2")
filename_O = os.listdir(path+"/O2")
for filename in filename_R:
    print(path+"/R2/"+filename)
    # load the image
    img = load_img(path+"/R2/"+filename)

    #print(type(img))
    #print(img.format)
    #print(img.mode)
    print(img.size)
    img.show()


     # convert to numpy array
    img_array = img_to_array(img)

    X = np.append (X,img_array)
    category = filename.split('.')[0]
    y = np.append(y,[0])

for filename in filename_O:
    print(path+"/O2/"+filename)
    # load the image
    img = load_img(path+"/O2/"+filename)

    #print(type(img))
    #print(img.format)
    #print(img.mode)
    print(img.size)
    img.show()


     # convert to numpy array
    img_array = img_to_array(img)

    X = np.append (X,img_array)
    category = filename.split('.')[0]
    y = np.append(y,[1])
```

```python
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense, BatchNormalization
```

```python
modelX = Sequential()
modelX.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(200, 200, 3)))
modelX.add(MaxPooling2D())
modelX.add(Dropout(0.2))

modelX.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
modelX.add(MaxPooling2D())
modelX.add(Dropout(0.2))

modelX.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
modelX.add(MaxPooling2D())
modelX.add(Dropout(0.2))

modelX.add(Flatten())
modelX.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
modelX.add(Dropout(0.5))
modelX.add(Dense(1, activation='sigmoid'))
# # compile model
opt = 'sgd'
modelX.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])



# modelX.add(Flatten())
# modelX.add(Dense(256))
# modelX.add(Activation("relu"))
# modelX.add(Dropout(0.5))
# modelX.add(Dense(64))
# modelX.add(Activation("relu"))
# modelX.add(Dropout(0.5))
# #model.add(Dense(numberOfClass)) # output
# modelX.add(Dense(1, activation='sigmoid'))
# modelX.add(Activation("sigmoid"))

# # modelX.compile(loss = "binary_crossentropy",
# #               optimizer = "adam",
# #               metrics = ["accuracy"])

# opt = 'sgd'  # Use string identifier for SGD optimizer
# modelX.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
historyD = modelX.fit(X_train, y_train, batch_size=50, epochs=20,validation_data = (X_test,y_test),callbacks =None)
```

```python
test_loss, test_accuracy = modelX.evaluate(X_test,  y_test, verbose=2)

print('\nTest accuracy:', test_accuracy)
```

```python
print(historyD.history.keys())
plt.figure(figsize=(9,6))
plt.plot(historyD.history['accuracy'])
plt.plot(historyD.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc = 'upper left')
plt.show()
```
